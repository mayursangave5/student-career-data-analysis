```python
import pandas as pd


#  LOAD DATA

df = pd.read_csv('../data/row data/student_placement_career_success_dataset.csv')

print("Data Loaded Successfully!")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns\n")

# DATA QUALITY ASSESSMENT 

print("=" * 50)
print(" MISSING VALUES BEFORE CLEANING:")
print("=" * 50)
missing = df.isnull().sum()
missing = missing[missing > 0]
print(missing)

print(f"\n Duplicate Rows: {df.duplicated().sum()}")


#  DATA CLEANING

print("\n" + "=" * 50)
print(" CLEANING DATA...")
print("=" * 50)

# Fix missing NUMERIC columns with median
numeric_cols = ['GitHub_repos', 'mock_interview_score', 'sleep_hours']
for col in numeric_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    print(f"✔ '{col}' filled with median = {median_val}")

# Fix missing TEXT columns with mode
text_cols = ['company_type', 'work_mode']
for col in text_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
    print(f"✔ '{col}' filled with mode = '{mode_val}'")

# Remove duplicates (if any)
df = df.drop_duplicates()
print(f"✔ Duplicates removed")


# FEATURE ENGINEERING (Extra Column)

# Create a new useful column - skill score combining multiple skills
df['overall_skill_score'] = (
    df['DSA_problems_solved'] * 0.3 +
    df['resume_score'] * 0.2 +
    df['communication_skills'] * 0.2 +
    df['aptitude_score'] * 0.2 +
    df['mock_interview_score'] * 0.1
).round(2)

print("✔ New column 'overall_skill_score' created")


# VERIFY (AFTER CLEANING)

print("\n" + "=" * 50)
print(" MISSING VALUES AFTER CLEANING:")
print("=" * 50)
print(df.isnull().sum().sum(), "missing values remaining")

print("\nCLEANED DATA SAMPLE:")
print(df.head(3))


# SAVE CLEANED DATA

df.to_csv('../data/Cleaned_data/cleaned_dataset.csv', index=False)
print("\n Cleaned dataset saved as 'cleaned_dataset.csv'")
print(f"Final Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```
